```python
##Question2
#European Binomial Option

import numpy as np


s0 = 1000   #Stock Price
N = 10000   #Number of trial
K = 1500    # strike price
u = 2       #up
d = 1/u     #down
r = 0.05    #interest
p = 0.5     #probability of each trial
C = 0
P = 0
T = 1       #year

for i in range(N):
    W = np.random.binomial(1, p)
    if W == 0:
        S = s0*d
        s0 = S
    else:
        S = s0*u
        s0 = S
        
    C += max(0, S - K)
    P += max(0, K - S)
    
call_price = (C/N)*np.exp(-r*T)
put_price = (P/N)*np.exp(-r*T)

print('Call price =' , call_price)
print('Put price =' , put_price)
```
